```python
copy = xrds.sel(observations=np.isin(xrds.individual, blood.individual))
# re-index blood xarray by individuals
blood_indivs = blood.rename_dims(observations="individual").set_index(individual="individual").set_coords("individual")
# rename variables
blood_indivs = blood_indivs.rename({k: k + "_blood" for k in blood_indivs.keys()})
# broadcast-select individuals corresponding to each observation in `copy`
blood_indivs = blood_indivs.reset_coords(drop=True).sel(individual=copy.individual)
# assign '*_blood' - variables to `copy`
copy = copy.assign(blood_indivs)
copy
```

```python
def group_counts():
    for subtissue, st_data in copy.groupby("subtissue"):
        for other_grp, other_data in st_data.groupby("hilo_padj"):
            for blood_grp, blood_data in other_data.groupby("hilo_padj_blood"):
                yield (
                    subtissue, # subtissue
                    other_grp, # group in tissue
                    blood_grp, # corresponding group in blood
                    np.size(blood_data.observations), # number of elements having that hilo_padj group
                    np.sum(blood_data.missing.values), # number of elements with unknown gene expression in tissue
                    np.sum(blood_data.missing_blood.values), # number of elements with unknown gene expression in blood
                )
group_counts = list(tqdm.tqdm(group_counts()))
group_counts = pd.DataFrame(group_counts, columns=["subtissue", "subtissue_expression_type", "blood_expression_type", "count", "n_missing_subtissue", "n_missing_blood"])
group_counts["subtissue_expression_type"] = pd.Categorical.from_codes(group_counts["subtissue_expression_type"].values + 1, categories=["underexpressed", "normal", "overexpressed"])
group_counts["blood_expression_type"] = pd.Categorical.from_codes(group_counts["blood_expression_type"].values + 1, categories=["underexpressed", "normal", "overexpressed"])
group_counts
```

```python
group_counts = group_counts.set_index(["subtissue", "blood_expression_type", "subtissue_expression_type"]).unstack("subtissue", fill_value=0).T
group_counts
```

```python
def ecdf(data):
    """ Compute ECDF """
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return(x,y)

x, y = ecdf(np.random.choice(n_values_per_obs.values.flatten(), 10000))

px.scatter(
    pd.DataFrame(dict(x=x, y=y)), 
    x="x", 
    y="y", 
    title="ECDF of the number of subtissues each gene is expressed in",
    labels={
        "x": "number of subtissues a gene is expressed in",
        "y": "ECDF",
    },
    render_mode="webgl"
)
```
