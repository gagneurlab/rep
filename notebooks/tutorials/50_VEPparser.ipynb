## parser

```python


import pandas as pd
outputDf=pd.DataFrame(index=["subject","gene",""])

def cacheLine(line):
    global lineCache
    lineCache.append(line)
    
def processCacheNrestartWith(line: dict):

    processCache()
    restartCache(line) 

def processCacheNrestartWith_old(line: dict):

    processCache_old()
    restartCache(line) 
    
def processCache():

    global lineCache
    global outWriter

    #continue at the start of the file
    if len(lineCache)==0: return 
    
    #concert to dataframe and reset list
    df1Gene1Indiv=pd.DataFrame(lineCache)
    lineCache=[]
 
    #create all the dicts for all the gene/subject-pairs
    dictList={}
    #grouping by gene/subject pair
    instanceGrouped=df1Gene1Indiv.groupby([f_gene,f_subject],sort=False) 
    for group in instanceGrouped:   # instanceGrouped.groups.keys(): is SORTED again! ->messes up order in frozenDict... -> would not keep input-order
        keys=group[0]
        idKey=keys2str(keys)
        dictList[idKey]={"gene":keys[0],"subject":keys[1]}
    
    
    #c_CONSEQUENCE_[hetero/homo] -> homo: only 1.3% of cases...
    consCounts=df1Gene1Indiv.groupby([f_gene,f_subject,"Consequence","GT"])["Consequence"].count()
    for keys, count in consCounts.iteritems():
        idKey=keys2str(keys[0:2])
        fieldKey=keys2str(keys[2:])
        dictList[idKey]["c_consequence_"+fieldKey]=count
        
    #c_CONSEQUENCE_[hetero/homo] -> homo: only 1.3% of cases...
    impactCounts=df1Gene1Indiv.groupby([f_gene,f_subject,"IMPACT","GT"])["IMPACT"].count()
    for keys, count in impactCounts.iteritems():
        idKey=keys2str(keys[0:2])
        fieldKey=keys2str(keys[2:])
        dictList[idKey]["c_impact_"+fieldKey]=count
    
    #[polyphen|sift|_[avg|max] # SIFT most-serious is 0, not 1 !! -) need min
    #convert text values to probs
    df1Gene1Indiv["SIFT"]=df1Gene1Indiv["SIFT"].apply(extractSiftPolyProb)
    df1Gene1Indiv["PolyPhen"]=df1Gene1Indiv["PolyPhen"].apply(extractSiftPolyProb)   
    #TODO actually use scores...
    values=instanceGrouped["SIFT"].min()
    for keys, val in values.iteritems():
        idKey=keys2str(keys[0:2])
        dictList[idKey]["sift_min"]=val
    values=instanceGrouped["SIFT"].mean()
    for keys, val in values.iteritems():
        idKey=keys2str(keys[0:2])
        dictList[idKey]["sift_avg"]=val   
        
    values=instanceGrouped["PolyPhen"].max()
    for keys, val in values.iteritems():
        idKey=keys2str(keys[0:2])
        dictList[idKey]["polyphen_max"]=val
    values=instanceGrouped["PolyPhen"].mean()
    for keys, val in values.iteritems():
        idKey=keys2str(keys[0:2])
        dictList[idKey]["polyphen_avg"]=val
    
    
    
    
    
    # LoF_[LC|HC] (ignoring NA) -> this is only 3% and 0.3% of cases
    lofCounts=df1Gene1Indiv.groupby([f_gene,f_subject,"LoF"])["LoF"].count()
    
    for keys, count in lofCounts.iteritems():
        idKey=keys2str(keys[0:2])
        fieldKey=keys2str(keys[2:])
        dictList[idKey]["c_lof_"+fieldKey]=count
        
        

    outWriter.writerows(dictList.values())
    
    return 
```

```python
   
def restartCache(line):
    global idTuple
    global lineCache
    
    idTuple=(line[f_subject],line[f_gene])
    lineCache=[line]
    
def extractSiftPolyProb(string):
    """e.g. polyphen-field: probably_damaging(0.993)"""
    if string=="NA": return np.nan #NA
    try:
        return float(string.split("(")[1][:-1]) #get string after "(", remove last letter...
    except:
            print(string)
    

    

def keys2str(keys):
    return "-".join(keys)
    
```

```python
def filterDict(largeDict):
    wantedKeys=[f_subject,f_gene,"IMPACT","Consequence","GT","SIFT","PolyPhen","LoF"]
    return { wKey:largeDict[wKey] for wKey in wantedKeys}
```

```python
%%time
#%%prun -s cumulative
#emergency exit: killall python --user reinharj

"""
for each line:
check if new gene or new indiv
if not, add to variant-collection
if yes: analyse cached variants, reset variant-collection

analyse cached variants:
    transform to dataframe, then group and analyze...
    output format: Gene count_<feature_type>_[hetero|homo] count_impact_[hetero|homo] [polyphen|sift|_[avg|max] LoF_[LC|HC]
"""
realVEP="/s/project/rep/predictions/vep/sortedVEPoutput.tsv"
outFile="/s/project/rep/predictions/vep/aggregatedVEP.tsv"
vepRaw=open(realVEP,"r",8000000)
outRaw=open(outFile,"w",8000000)
processLimit=100000

idTuple=("indiv","gene")
lineCache=[]

vepDict=csv.DictReader(vepRaw,delimiter="\t")
outWriter=csv.DictWriter(outRaw,fieldnames=header)
outWriter.writeheader()
c=0
oldSave=0
for line in vepDict:
    c+=1
    if c%100000==0: print(c)
    smallDict=filterDict(line) #reduces time needed from 117 to 52 s for 10k lines... (slow algo)
    #continue until limit reached AND differnt gene/subject combo
    if len(lineCache)< processLimit or (smallDict[f_subject]== lineCache[-1][f_subject] and smallDict[f_gene]==lineCache[-1][f_gene]): 
            cacheLine(smallDict)
    else: 
        processCacheNrestartWith(smallDict)
        
processCache() #end fence-post
            
vepRaw.close()
outRaw.close()
```
